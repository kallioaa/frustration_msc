```python
from environments.fronzenlake import get_frozenlake_env
from agents.sarsa_td0 import SarsaTD0Agent, SarsaTD0Config
from experiments.evaluate_win_rate import evaluate_win_rate
from plots.plot_episode_returns import plot_episode_returns
```

```python
env = get_frozenlake_env(map_name="4x4", is_slippery=True)
```

```python
# create SARSA(0) agent with default configuration

sarsa_td0_agent = SarsaTD0Agent(
    config=SarsaTD0Config(epsilon=0.1)
)

# train the agent
returns = sarsa_td0_agent.train(env, num_episodes=10000)
```

```python
# evaluate the agent
win_rate, total_wins = evaluate_win_rate(env, sarsa_td0_agent, num_episodes=1000)
print(f"Win rate: {win_rate:.2%} ({total_wins} wins out of 1000 episodes)")
```

```python
# plot episode returns
plot_episode_returns(returns=returns)
    
```
